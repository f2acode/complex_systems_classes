```python
# exercise pi2
import numpy as np
from matplotlib import pyplot as plt
import math

x0 = 0
y0 = 1
xf = 10
n = 101

deltax = (xf - x0) / (n - 1)
x = np.linspace (x0 ,xf ,n)
y = np.zeros([n])
y [0] = y0

for i in range (1, n):
    y[i] = deltax *(math.sqrt(math.cos(x[i - 1]) ** 2 + math.sin(y[i - 1]) ** 2)) + y[i - 1]

plt.plot(x ,y ,'or')
plt.xlabel("Value of x")
plt.ylabel("Value of t")
plt.title("Approximate Solution by Euler’s Method")
plt.show ()
```

```python
import numpy as np
from matplotlib import pyplot as plt
import math

x0 = 0
y0 = 1
xf = 10
n = 101

deltax = (xf - x0) / (n - 1)
x = np.linspace (x0 ,xf ,n)
y = np.zeros([n])
y [0] = y0

for i in range (1, n):
    y[i] = deltax *(math.cos(x[i - 1] * y[i - 1])) + y[i - 1]

plt.plot(x ,y ,'or')
plt.xlabel("Value of x")
plt.ylabel("Value of y")
plt.title("Approximate Solution by Euler’s Method")
plt.show ()
```

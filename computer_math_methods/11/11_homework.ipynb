$$\Delta(a, b) := \sum_{i=1}^{n}[yi - (ax_i + b)]^2$$

Derivando em relação à $a$ e em relação à $bx$:

$$\frac{\partial M}{\partial a} = 2 \sum_{i=1}^{n}(y_i - a - bx_i)(-1) = 0$$

$$\frac{\partial M}{\partial b} = 2 \sum_{i=1}^{n}(y_i - a - bx_i)(-x_i) = 0$$

Multiplicando pelo termo obtido após a derivação e cancelando o fator 2 (já que igualamos à zero, então não fará diferença o 2):

$$= \sum_{i=1}^{n}(-y_i + a + bx_i) = 0$$

$$= \sum_{i=1}^{n}(-x_iy_i +ax_i + bx_i^2) = 0$$

Separando as somatórias:

$$= \sum_{i=1}^{n}(-y_i)  + \sum_{i=1}^{n}a + \sum_{i=1}^{n}bx_i = 0$$

$$= \sum_{i=1}^{n}(-x_iy_i)  + \sum_{i=1}^{n}ax_i + \sum_{i=1}^{n}bx_i^2 = 0$$

Passando as somatórias que não envolvem a ou b para o segundo membro:

$$= \sum_{i=1}^{n}a + \sum_{i=1}^{n}bx_i = \sum_{i=1}^{n}y_i$$

$$= \sum_{i=1}^{n}ax_i + \sum_{i=1}^{n}bx_i^2 = \sum_{i=1}^{n}x_iy_i$$

Colocando a e b em evidência:

$$= a\sum_{i=1}^{n} + b\sum_{i=1}^{n}x_i = \sum_{i=1}^{n}y_i$$

$$= a\sum_{i=1}^{n}x_i + b\sum_{i=1}^{n}x_i^2 = \sum_{k=1}^{n}x_iy_i$$

```python
import matplotlib.pyplot as plt
import numpy as np

# make the chart bigger to see all points
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 12), dpi=150)

file = open('data-l11-2019.txt', 'r')

n, x_dots, y_dots = 0, [], []

for line in file:
    x_dots.append(float(line.split()[0]))
    y_dots.append(float(line.split()[1]))
    n += 1
    
file.close()

sum_one, sum_x, sum_power_x, sum_y, sum_x_y = 0, 0, 0, 0, 0

for i in range(n):
    sum_one += 1
    sum_x += x_dots[i]
    sum_power_x += x_dots[i] ** 2
    sum_y += y_dots[i]
    sum_x_y += x_dots[i] * y_dots[i]

a = [[sum_one, sum_x], [sum_x, sum_power_x]]
b = [sum_y, sum_x_y]

r = np.linalg.solve(a, b)

# write the result
student_file = open('felipe-augusto-de-almeida.txt', 'w')
student_file.write(f'{r[0]} {r[1]}')
student_file.close()

y_linear = [(r[1]*x)+r[0] for x in x_dots]

plt.title('Linear Regression', fontsize=20)
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y$', fontsize=20)
plt.plot(x_dots, y_dots, 'bo')
plt.plot(x_dots, y_linear, '-r')
plt.show()
```

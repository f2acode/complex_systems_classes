```python
# pi1
import networkx as nx
import matplotlib.pyplot as plt
import random

N = 60#480
G = nx.Graph()

for i in range(N):
    G.add_node(i)
    
for i in range(N):
    if i == len(range(N)) - 1:
        G.add_edge(i, 0)
        print(i)
    else:
        G.add_edge(i, i+1)

setd = G.degree()
nx.draw(G, node_size=100, alpha=1.0, node_color='red', with_labels=False)
plt.show()
```

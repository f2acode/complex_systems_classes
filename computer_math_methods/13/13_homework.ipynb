```python
import numpy as np
import math
import matplotlib.pyplot as plt

file = np.loadtxt('data-scx5000-l13-2019-1.txt')
file[0]

n = len(file)

# (i)
x, y = [], []
for i in file:
    x.append(i[0])
    y.append(i[1])
    
plt.scatter(x, y)
plt.show()

# (ii)
ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')
plt.ylim((0.0000001,10))
plt.scatter(x, y)
plt.show()

# (iii)
xi, xi2, yi, xiyi = 0, 0, 0, 0
for i in file:
    xi += math.log(i[0])
    xi2 += (math.log(i[0]) ** 2)
    
    if(i[1] != 0):
        yi += math.log(i[1])
        xiyi += math.log(i[0])*math.log(i[1])
        
    x.append(i[0])
    y.append(i[1])

a=(n * xiyi - xi * yi)/(n * xi2 - xi ** 2)
b=(yi - a * xi)/(n)
print(xi, xi2, yi, xiyi)
print(a, b)

# (iv)
x1, y1 = [],[]
xl = np.linspace(min(x), max(x), 100)
yl = a * xl ** b

ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')

plt.scatter(x, y)
plt.plot(xl, yl, '-r')
plt.legend(['y=Ax^b','dots'])

plt.show()
```

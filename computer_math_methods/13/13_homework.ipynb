```python
import numpy as np
import math
import matplotlib.pyplot as plt

file = np.loadtxt('data-scx5000-l13-2019-1.txt')
file[0]

n = len(file)

# (i)
x, y = [], []
Sx, Sx2, Sy, Sxy = 0, 0, 0, 0

for i in file:
    x.append(i[0])
    y.append(i[1])
    
plt.scatter(x, y)
plt.show()

# (ii)

ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')
plt.ylim((0.00000001,10))
plt.scatter(x, y)
plt.show()

# (iii)
Sx,Sx2,Sy,Sxy = 0, 0, 0, 0
for i in file:
     Sx+=math.log(i[0])
     Sx2+=(math.log(i[0])**2)
     if(i[1]!=0):
         Sy+=math.log(i[1])
         Sxy+=math.log(i[0])*math.log(i[1])
     x.append(i[0])
     y.append(i[1])

a=(n*Sxy-Sx*Sy)/(n*Sx2-Sx**2)
b=(Sy-a*Sx)/(n)
print(Sx,Sx2,Sy,Sxy)
print(a,b)

# (iv)
x1,y1=[],[]
xl = np.linspace(min(x),max(x),100)
yl = a*xl**b

ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')

plt.scatter(x, y)
plt.plot(xl, yl, '-r')
plt.legend(['y=Ax^b equation line','Points'])

plt.show()

out=[(a,b)]
np.savetxt('felipe.txt', out, delimiter=' ')
```

## Enrenfest run

```python
import random
import matplotlib.pyplot as plt

#L = int(input('Initial number of balls (left):'))
#R = int(input('Initial number of balls (right):'))
#Ttot = int(input('Total time:'))

'''
prob sorteio da esquerda
L/L+R
prob sorteio da direita
R/L+R
'''
def enrenfestUrn(L, R, Ttot):
    t_list, left_list = [0], [L]
    for t in range(Ttot):
        r = random.uniform(0.0, 1.0)
        if r < L/float(L + R):
            L = L - 1
            R = R + 1
        else:
            L = L + 1
            R = R - 1
        t_list.append(t + 1)
        left_list.append(L)

    '''
    Plot a graph

    matplotlib.pyplot as plt imported
    Lists t_list and left_list of t and L
    '''

    plt.plot(t_list, left_list) # Graph left_list x t_list
    plt.xlabel('Time ') # Label of x axis
    plt.ylabel('Balls (left)') # Label of y axis
    plt.ylim([0, L + R]) # Range of y axis
    plt.title('Ehrenfest urn') # Title
    plt.show()
```

```python
enrenfestUrn(10, 0, 100)
```

```python
enrenfestUrn(10, 0, 5000)
```

### But now we create a new function doing the simulation many times

We need to remember to play many times with our data to ensure we can say something about what we are studying

```python
def enrenfestUrnV2(L, R, Ttot, run):
    t_list, left_list = [0], [L]
    for n in range(run):
        for t in range(Ttot):
            r = random.uniform(0.0, 1.0)
            if r < L/float(L + R):
                L = L - 1
                R = R + 1
            else:
                L = L + 1
                R = R - 1
            t_list.append(t + 1)
            left_list.append(L)

    plt.plot(t_list, left_list) # Graph left_list x t_list
    plt.xlabel('Time ') # Label of x axis
    plt.ylabel('Balls (left)') # Label of y axis
    plt.ylim([0, L + R]) # Range of y axis
    plt.title('Ehrenfest urn') # Title
    plt.show()
```

```python
enrenfestUrnV2(10, 0, 100, 1)
```

```python
enrenfestUrnV2(10, 0, 100, 10)
```

```python
enrenfestUrnV2(10, 0, 100, 1000)
```

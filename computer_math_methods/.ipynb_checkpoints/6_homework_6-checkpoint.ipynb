```python
#1
import random
import matplotlib.pyplot as plt

N = 30
T = 100
L, C, R = 0, 0, N
urnL, urnC, urnR = [], [], []
total = []
    
for t in range(T):
    if L != 0:
        L -= 1
        r = random.uniform(0.0, 1.0)
        if r > 0.5:
            R += 1
        else:
            C += 1
        
    if C != 0:
        C -= 1
        r = random.uniform(0.0, 1.0)
        if r > 0.5:
            L += 1
        else:
            R += 1
    
    if R != 0:
        R -= 1
        r = random.uniform(0.0, 1.0)
        if r > 0.5:
            L += 1
        else:
            C += 1
    
    total.append(t + 1)
    urnL.append(L)
    urnC.append(C)
    urnR.append(R)
    

plt.plot(total, urnL)
plt.plot(total, urnR)
plt.plot(total, urnC)
plt.xlabel('Time')
plt.ylabel('Balls')
plt.title('Ehrenfest urn')
plt.show()
```

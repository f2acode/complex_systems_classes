```python
#1
import random
import matplotlib.pyplot as plt

R = 1
N = 30
T = 100
Lvalue, Cvalue, Rvalue = 0, 0, N
urnL, urnC, urnR = [], [], []
total = []

def run():
    global Lvalue, Cvalue, Rvalue
    for rr in range(R):
        for t in range(T):
            if Lvalue != 0:
                Lvalue -= 1
                r = random.uniform(0.0, 1.0)
                if r > 0.5:
                    Rvalue += 1
                else:
                    Cvalue += 1

            if Cvalue != 0:
                Cvalue -= 1
                r = random.uniform(0.0, 1.0)
                if r > 0.5:
                    Lvalue += 1
                else:
                    Rvalue += 1

            if Rvalue != 0:
                Rvalue -= 1
                r = random.uniform(0.0, 1.0)
                if r > 0.5:
                    Lvalue += 1
                else:
                    Cvalue += 1

            if rr == 0:
                urnL.append(Lvalue)
                urnC.append(Cvalue)
                urnR.append(Rvalue)
                total.append(t + 1)
            else:
                urnL[t] = (urnL[t] + Lvalue)/2
                urnC[t] = (urnC[t] + Cvalue)/2
                urnR[t] = (urnR[t] + Rvalue)/2
    

    plt.plot(total, urnL)
    plt.plot(total, urnR)
    plt.plot(total, urnC)
    plt.xlabel('Time')
    plt.ylabel('Balls')
    plt.title('Ehrenfest urn with R = 1')
    plt.show()
    
run()

R = 30

run()
```

```python
import matplotlib.pyplot as plt
import random
import numpy as np

N = 100
total_time = 100

voters = [0 for i in range(N)]
time, m = [0], [0]

for i in range(N):
    r = random.uniform(0.0, 1.0)
    if r < 0.5:
        voters[i] = -1
    else:
        voters[i] = 1

m[0] = np.mean(voters)

for t in range(1, total_time + 1):
    i = random.randint(0, N-1)
    j = random.randint(0, N-1)
    voters[i] = voters[j]
    m.append(np.mean(voters))
    time.append(t)
    
plt.plot(time, m)
plt.xlabel("time")
plt.ylabel("")
plt.xlim(0, total_time)
plt.ylim(-1, 1)
plt.title("Voters model")
plt.show()
```

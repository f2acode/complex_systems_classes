### Analytical

$$y'(x) = \frac{e^{y(x)}}{x}$$

$$\frac{y'(x)}{e^{y(x)}} = \frac{1}{x}$$

$$y'(x)e^{-y(x)} = \frac{1}{x}$$

$$\int{e^{-y(x)}} \frac{dy(x)}{dx}dx = \int{\frac{1}{x}}dx$$

$$-e^{-y(x)} = log(x) + c_1 \rightarrow \text{Calculando as integrais}$$ 

$$log^{-e^{y(x)}} = log^{log(x) + c_1}$$

$$-y(x) = (log(x) + c_1) log$$

$$y(x) = -log(-log(x) + c_1):$$

$$-log(c_1) = 0$$

$$\text{Encontrando } c_1 \text{ usando as condições iniciais}$$

$$y(1) = 0$$

$$y(2) = -log(-log(x) + c_1)$$

$$0 = -log(-log(0)+c_1)$$

$$-log(c_1) = 0$$

$$c_1 = 1$$

$$\text{Substituindo } c_1 = 1 \text{ em}$$

$$y(x) = -log(-log(x) + c_1)$$

$$y(x) = -log(-log(x) + 1)$$

```python
import matplotlib.pyplot as plt
import math, warnings
import numpy as np
```

```python
def analytical(x):
    return -np.log(1 - np.log(x))

def euler(x0, y0, h, x_end):
    x, y = [], []
    x.append(x0)
    y.append(y0)
    while x0 < x_end:
        x1 = x0 + h
        y1 = y0 + (h/x0) * np.e ** y0
        x0, y0 = x1, y1
        x.append(x0)
        y.append(y0)
    return x, y

def runge_kutta(x0, y0, x, x_end):
    x, y = [], []
    while x0 < x_end:
        k1 = (1/x0) * np.e ** y0
        k2 = (1/(x0 + h)) * np.e ** (y0 + h)
        
        x1 = x0 + h
        y1 = y0 + (h/2) * (k1 + k2)
        
        x0, y0 = x1, y1
        x.append(x0)
        y.append(y0)
    return x, y
```

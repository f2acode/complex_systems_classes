```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
import math

G=nx.Graph() #graph
N=1000 #vertices
t=[0]
T=N*(N-1) #times
p=5/N

for i in range(0,N):
    G.add_node(i)
    
for time in range(0,T): # (ii), (iv)

    i=random.randint(0,N-1) #first vertice
    j=random.randint(0,N-1) #second vertice
    t.append(time)

    prob=random.uniform(0.0,1.0) #connect with probability p
    if prob < p: # (iii)
        G.add_edge(i,j)

        # draw network graph
import matplotlib.pyplot as plt
nx.draw(G,node_size=50) # , with_labels=True,font_size=8
```

```python
import numpy as np
import math

degrees = [val for (node, val) in G.degree()]# degree
plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.plot()

def fp(k): # poisson function
    u=p * (N-1) * 2
    K=[]

    for i in range(0,len(k)):
        K.append(((math.exp(-u) * u**k[i]) / (math.factorial(k[i])) )*N)
    return (K)

bins=20 # histogram
plt.rcParams['figure.figsize'] = [16, 8]

counts, bins = np.histogram(degrees,bins=bins)
plt.hist(degrees,bins=bins)

k = np.arange(min(bins), max(bins), 1)
plt.plot(k,fp(k), 'r')

plt.legend(['Poissonian degree distribution','Erdos-Renyi degree distribution'])
```

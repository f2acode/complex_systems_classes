### "Erdos-RÃ©nyi" graph

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

N = 10 ** 2
p = 0.06

G = nx.Graph()

for i in range(N):
    G.add_node(i)
    
for i in range(N):
    for j in range(i + 1, N):
        if random.uniform(0.0, 1.0) < p:
            G.add_edge(i, j)
            
nx.draw(G, node_size=100, alpha=1.0, node_color='red', with_labels=False)
plt.show()
```

### "Erdos-RÃ©nyi"

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

N = 10 ** 2
p = 0.06

G = nx.Graph()

for i in range(N):
    G.add_node(i)
    
for i in range(N):
    for j in range(i + 1, N):
        if random.uniform(0.0, 1.0) < p:
            G.add_edge(i, j)

setd = G.degree()
nx.draw(G, node_size=[val * 100 for (node, val) in setd], alpha=1.0, node_color='red', with_labels=False)
plt.show()
```

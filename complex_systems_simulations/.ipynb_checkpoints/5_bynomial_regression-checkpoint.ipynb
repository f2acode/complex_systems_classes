## Bynomial regression

1. Gerar números (x) aleatórios 
2. Definir a função de probabilidade
3. Gerar errors (feito em distribuição normal)
4. Pegar dados da distribuição binomial e fazer regressão

```python
import random, math, warnings
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
# this is to avoid display a warning: FutureWarning: Using a non-tuple sequence for multidimensional indexing is deprecated use `arr[tuple(seq)]` instead of `arr[seq]`

N = 10
x_values, y_values = np.array([]), np.array([])
plt.figure(figsize=(20,10))

def prob(x):
    return math.factorial(N)/(math.factorial(x) * math.factorial(N - x))

def generate_error():
    return random.uniform(0, N)
    
for i in range(N):
    x_values = np.append(x_values, i)
    y_values = np.append(y_values, int(prob(i) + generate_error()))

sns.barplot(x_values, y_values)

x_values = x_values.reshape((-1, 1))
model = LinearRegression().fit(x_values, y_values)
r_sq = model.score(x_values, y_values)
y_pred = model.predict(x_values)

plt.plot(x_values, y_pred, color="red")
plt.show()
```
